```python
import matplotlib.pyplot as plt
import numpy as np

import pyxdf
```

```python
data, header = pyxdf.load_xdf("sub-P001_ses-S001_task-Default_run-001_eeg.xdf")
```

```python
stream_list = [stream['info']['name'][0] for stream in data]
stream_list
```

```python
index = stream_list.index('Unity.Position.Ball')
```

```python
stream = data[index]
```

```python
stream.keys()
```

```python
stream['info']['desc'][0]['channels'][0]['channel']
```

```python
plt.plot(stream['time_stamps']-stream['time_stamps'][0], stream['time_series'])
plt.legend([item['label'][0] for item in stream['info']['desc'][0]['channels'][0]['channel']], loc='upper left')
plt.title(stream['info']['name'][0])
plt.xlabel("Timestamps")
plt.ylabel(stream['info']['type'][0])
```

```python
idx1 = stream_list.index('Unity.Position.Player1')
idx2 = stream_list.index('Unity.Position.Player2')
idx3 = stream_list.index('Unity.GameManagerState')
```

```python
strm1 = data[idx1]
strm2 = data[idx2]
strm3 = data[idx3]
```

```python
plt.plot(strm1['time_stamps']-strm1['time_stamps'][0], strm1['time_series'][:, 1], label=strm1['info']['name'][0])
plt.plot(strm2['time_stamps']-strm2['time_stamps'][0], strm2['time_series'][:, 1], label=strm2['info']['name'][0])
plt.plot(strm3['time_stamps']-strm3['time_stamps'][0], strm3['time_series'][:, 1], label=strm2['info']['name'][0])
plt.legend([strm1['info']['name'][0], strm2['info']['name'][0]], loc='upper left')
plt.title("Position")
plt.xlabel("Timestamps")
plt.ylabel(stream['info']['type'][0])
plt.show()
```
